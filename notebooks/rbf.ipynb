```python
from collections.abc import Callable

import matplotlib.pyplot as plt
import numpy as np

type Array1d = np.ndarray[tuple[int], np.dtype[np.number]]
```

```python
def get_tanh_rbf(eps: float) -> Callable:
    def _tanh_rbf(x: Array1d) -> Array1d:
        r = np.linalg.norm(x)
        return r * np.tanh(eps * r)
    return _tanh_rbf
```

```python
xs = np.linspace(0.0, 2.0, 50)

plt.plot(xs, xs, label="x")
plt.plot(xs, list(map(get_tanh_rbf(0.1), xs)), label="eps=0.1")
plt.plot(xs, list(map(get_tanh_rbf(0.5), xs)), label="eps=0.5")
plt.plot(xs, list(map(get_tanh_rbf(1.0), xs)), label="eps=1.0")
plt.plot(xs, list(map(get_tanh_rbf(2.0), xs)), label="eps=2.0")

plt.legend()
plt.show()
```
